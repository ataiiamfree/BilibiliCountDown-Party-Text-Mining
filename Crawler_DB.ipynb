```python
import jieba
comment_Bilibili = ' · '.join(df_Bilibili['content'])
comment_Jiangsu = ' · '.join(df_Jiangsu['content'])
comment_Hunan = ' · '.join(df_Hunan['content'])
comment_Dongfang = ' · '.join(df_Dongfang['content'])

comment_list_Bilibili = comment_Bilibili.split(' · ')
comment_list_Jiangsu = comment_Jiangsu.split(' · ')
comment_list_Hunan = comment_Hunan.split(' · ')
comment_list_Dongfang = comment_Dongfang.split(' · ')


tokenized_comment_Bilibili = list(jieba.lcut(comment_Bilibili))
tokenized_comment_Jiangsu = list(jieba.lcut(comment_Jiangsu))
tokenized_comment_Hunan = list(jieba.lcut(comment_Hunan))
tokenized_comment_Dongfang = list(jieba.lcut(comment_Dongfang))

print(tokenized_comment_Bilibili[0:30])
print("---------")
print(tokenized_comment_Jiangsu[0:30])
print("---------")
print(tokenized_comment_Hunan[0:30])
print("---------")
print(tokenized_comment_Dongfang[0:30])
```
